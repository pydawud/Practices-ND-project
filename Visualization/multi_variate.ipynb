```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('pokemon.csv')
df.head()
```

```python
df2 = pd.read_csv('fuel-econ.csv')
df2.head()
```

### Exercise 1:

```python
sns.scatterplot(data=df, x='defense',y='special-defense',hue='speed')
```

```python
#defense, special-defense and speed
plt.figure(figsize = [8,6])
plt.scatter(data = df, x = 'defense', y = 'special-defense',
            c = 'speed')
plt.colorbar(label = 'Speed')
plt.xlim(0,160)
plt.ylim(15,160)
plt.xlabel('Defense')
plt.ylabel('Special Defense')
```

## Adaptation of Bivariate for Multivariate

```python
sns.pointplot(data=df2, x='VClass', y='comb', hue='trans_type', ci='sd',linestyles='')
plt.xticks(rotation=60);
```
